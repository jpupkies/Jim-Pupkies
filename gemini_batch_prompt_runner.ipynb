```python
from google.genai import types

def send_prompt(prompt,
                model="gemini-2.0-flash",
                temperature=0.7,
                max_output_tokens=256):
    resp = client.models.generate_content(
        model=model,
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=temperature,
            max_output_tokens=max_output_tokens
        )
    )
    return resp.text

print(send_prompt("Write a short summary of the impact of AI in healthcare."))
```

```python
import pandas as pd
from google.genai import types

# Load prompts
df = pd.read_csv("prompts.csv")

# Prepare an empty list for responses
responses = []

# Process each row
for i, row in df.iterrows():
    prompt = row["prompt"]

    try:
        resp = client.models.generate_content(
            model="gemini-2.0-flash",
            contents=prompt,
            config=types.GenerateContentConfig(
                temperature=0.7,
                max_output_tokens=512
            )
        )
        text = resp.text

    except Exception as e:
        text = f"ERROR: {str(e)}"

    responses.append(text)

# Add responses to DataFrame
df["response"] = responses

# Save to CSV
df.to_csv("output.csv", index=False)

df  # Display the result
```

# Gemini Batch Prompt Runner

## Overview
This notebook demonstrates a batch AI automation workflow using the Gemini 2.0 LLM.  
Prompts are stored in a CSV (`prompts.csv`), sent to Gemini, and responses are saved in `output.csv`.

## How It Works
1. Load `prompts.csv` containing your prompts.
2. Iterate through each prompt and send it to Gemini using the Python SDK.
3. Save responses in `output.csv` for review or further use.

## Skills Demonstrated
- Python scripting
- CSV data processing
- Generative AI integration
- Error handling and output logging
- Reproducible workflow creation

## How to Run
1. Upload your `prompts.csv` in the same directory.
2. Set your Gemini API key in the Colab environment.
3. Run all cells to generate `output.csv`.
