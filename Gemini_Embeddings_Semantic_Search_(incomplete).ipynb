**Step 2: Configure Gemini API & Prepare Sample Data**

```python
import getpass
genai.configure(api_key=getpass.getpass("Enter your Gemini API key: "))

#Prepare a sample DataFrame:
df = pd.DataFrame({
    "text": [
        "Hello world!",
        "This is a test sentence.",
        "Gemini embeddings are cool."
    ]
})
```

**What this does:**

- Prompts you for your Gemini API key securely.

- Creates a small dataset of example text snippets in a Pandas DataFrame.

- This dataset will be used to generate embeddings in the next step.

**Step 3: Generate Embeddings for Your Text Data**

```python
# Step 3: Generate embeddings for sample data

from google.generativeai import embedding

def generate_embedding(text, model="gemini-embedding-001"):
    """
    Returns the embedding vector for a given text using Gemini embedding model.
    """
    response = embedding.embedContent(model=model, input=text)
    return response.data[0].embedding

# Apply to the DataFrame
df["embedding"] = df["text"].apply(generate_embedding)

# Preview embeddings
print(df.head())
```
