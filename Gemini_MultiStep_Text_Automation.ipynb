**Step 3-5 — Use fully simulated outputs**

```python
# Sample data
df = pd.DataFrame({
    "text": [
        "Alice wrote a blog post about AI on Dec 5th.",
        "Bob organized a team meeting for the marketing project tomorrow.",
        "Charlie found a bug in the software module last Friday."
    ]
})

# Pre-generated multi-step outputs
simulated_outputs = [
    {
        "summary": "Alice wrote a blog post about AI.",
        "rewritten": "Alice composed a professional blog post discussing AI.",
        "extracted": {
            "person": "Alice",
            "action": "wrote",
            "object": "blog post",
            "date/time": "Dec 5th"
        }
    },
    {
        "summary": "Bob organized a marketing team meeting.",
        "rewritten": "Bob scheduled a professional marketing team meeting.",
        "extracted": {
            "person": "Bob",
            "action": "organized",
            "object": "meeting",
            "date/time": "tomorrow"
        }
    },
    {
        "summary": "Charlie found a software bug last Friday.",
        "rewritten": "Charlie identified a bug in the software module last Friday.",
        "extracted": {
            "person": "Charlie",
            "action": "found",
            "object": "software bug",
            "date/time": "last Friday"
        }
    }
]

df["processed"] = simulated_outputs

# Expand JSON
df_expanded = pd.json_normalize(df["processed"].apply(lambda x: x["extracted"]))
df_final = pd.concat([df, df_expanded], axis=1).drop(columns=["processed"])

from IPython.display import display
display(df_final)
```
