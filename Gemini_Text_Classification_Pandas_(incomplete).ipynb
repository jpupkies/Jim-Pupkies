**Step 2 — Configure Gemini API Key**

```python
import getpass

# Prompt for API key securely
api_key = getpass.getpass("Enter your Gemini API key: ")

# Configure Gemini SDK
genai.configure(api_key=api_key)
```

**Step 3 — Prepare Sample DataFrame**

```python
import pandas as pd

# Create a simple DataFrame with text samples
df = pd.DataFrame({
    "text": [
        "Hello world!",
        "This is a test sentence.",
        "Large language models can classify text."
    ]
})

# Preview the DataFrame
df
```

✅ **Notes:**

1. pandas was already imported in Step 1.

2. Keep the dataset small and readable — perfect for demonstrating classification.

3. The column "text" will be used in Step 4 to apply the Gemini classification model.

```python
model = genai.GenerativeModel("gemini-1.5-flash")
print(model.generate_content("Reply with OK only.").text)
```

**Step 4 — Define Classification Function and Apply**

```python
import json

# Define a function to classify text using Gemini
def classify_text(text):
    prompt = f"""
Classify the following text into one of these categories:
- Greeting
- General
- Technical

Return ONLY valid JSON in this format:
{{"category": "<category>"}}

Text:
{text}
"""
    # Generate classification from the model
    response = model.generate_content(prompt)

    # Parse JSON output and return the category
    return json.loads(response.text)["category"]

# Apply the function to the DataFrame
df["category"] = df["text"].apply(classify_text)

# Preview results
df
```
